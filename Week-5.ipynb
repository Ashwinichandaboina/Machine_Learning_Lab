```python
mean=df['Age'].mean()
std=df['Age'].std()
num=df['Age'].isnull().sum()
val=np.random.normal(mean,std,num)
df.loc[df['Age'].isnull(),'Age']=val
print(df['Age'].isnull().sum())
```

```python
df['FamilySize']=df['Parch']+df['SibSp']+1
df['FamilySize']
```

```python
df['IsAlone']=df['FamilySize']==1
df['IsAlone']
```

```python
df['asCabin']=~df['Cabin'].isnull()
```

```python
df=df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=df.select_dtypes(include=['object']).columns
for col in cols:
    df[col]=le.fit_transform(df[col])
    
```

```python
df_categorical=df.select_dtypes(include=['object']).columns
d_categorical=df[df_categorical]
numeric=df.drop(columns=df_categorical)
final=pd.concat([df,d_categorical])
```

```python



x=df.drop(columns="Survived")
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='gini',max_depth=3)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df.dtypes
```

```python
from sklearn.tree import plot_tree
plot_tree(model,feature_names=x.columns,class_names=['Survived','Not Survived'],filled=True)
plt.show()
```

```python
plot_tree(model,feature_names=x.columns,class_names=['Survived','Not Survived'],filled=True,max_depth=3)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
